# Task#4: Data Cleaning/ Feature Engineering

```python
# Let's drop the date
df_alexa = df_alexa.drop(['date', 'rating'],axis=1)
```

```python
#display the dataset without the date
df_alexa
```

```python
# assign a variable named variation_dummies with numerical values for variation. 
variation_dummies = pd.get_dummies(df_alexa['variation'], drop_first = True)
```

```python
#display variation_dummies
variation_dummies
```

```python
# first let's drop the column
df_alexa.drop(['variation'], axis=1, inplace=True)
```

```python
# Now let's add the encoded column again
df_alexa = pd.concat([df_alexa, variation_dummies], axis=1)
```

```python

df_alexa
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
alexa_countvectorizer = vectorizer.fit_transform(df_alexa['verified_reviews'])
```

```python
alexa_countvectorizer.shape
```

```python
type(alexa_countvectorizer)
```

```python
print(vectorizer.get_feature_names())
```

```python
print(alexa_countvectorizer.toarray())  
```

```python
# first let's drop the column
df_alexa.drop(['verified_reviews'], axis=1, inplace=True)
reviews = pd.DataFrame(alexa_countvectorizer.toarray())
```

```python
# Now let's concatenate them together
df_alexa = pd.concat([df_alexa, reviews], axis=1)
```

```python
df_alexa
```

```python
# Let's drop the target label coloumns
X = df_alexa.drop(['feedback'],axis=1)
```

```python
# X = pd.DataFrame(alexa_countvectorizer.toarray())
```

```python
X
```

```python
y = df_alexa['feedback']
```

# Task #5: Model Building and Model Training

```python
#create a model to do a test split and train the model
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=5)
```

```python
#determine the shape of the x training model
X_train.shape
```

```python
#determine the shape of the x testing model
X_test.shape
```

```python
#determine the shape of the y training model
y_train.shape
```

```python
#determine the shape of the y testing model
y_test.shape
```

```python
#train the model using the training set 
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier

randomforest_classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy')
randomforest_classifier.fit(X_train, y_train)
```

# Task #6: Model Evaluation

```python
#create a confusion matrix wuth the training set
y_predict_train = randomforest_classifier.predict(X_train)
cm = confusion_matrix(y_train, y_predict_train)
```

```python
#create a heatmap
sns.heatmap(cm, annot=True)
```

```python
#print the classification report based on the training set
print(classification_report(y_train, y_predict_train))
```

```python
#create a confusion matrice with the testing set
y_predict = randomforest_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict)
```

```python
#create a heatmap with the testing set
sns.heatmap(cm, annot=True)
```

```python
# print a classification report based on the testing set
print(classification_report(y_test, y_predict))

#our model has a precision of 0.73 that 
#Our model has a precision of 0.89 so,when it predicts the actual label,it is correct at 89% of the time.
#Our model has a recall of 0.89, so out of the 89% positive data, our machine predicted the 89%
```

```python
# Let's drop the variations and add review length as a feature
df_alexa = pd.read_csv('https://raw.githubusercontent.com/niteen11/lagcc_data_analytics_micro_credential/master/Unit%204%20-%20Mahine%20Learning/amazon_alexa.tsv', sep='\t')
```

```python
X
```

```python
y = df_alexa['feedback']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier

randomforest_classifier = RandomForestClassifier(n_estimators = 300, criterion = 'entropy', random_state = 0)
randomforest_classifier.fit(X_train, y_train)
y_predict = randomforest_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
print(classification_report(y_test, y_predict))
```
